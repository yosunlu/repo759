```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Array sizes from 2^10 to 2^30
array_sizes = [2**i for i in range(10, 31)]

# Execution times extracted from the output
execution_times = [
    0.007694, 0.006051, 0.015912, 0.020081, 0.05627, 0.121368, 0.287577,
    0.507948, 1.09843, 2.10178, 4.15731, 8.17888, 12.8439, 33.4913, 52.0141,
    97.8696, 195.907, 391.703, 797.381, 1576.21, 3152.55
]
```

```python
# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(array_sizes, execution_times, marker='o', linestyle='-', color='b')

# Setting axis labels and title
plt.xlabel("Array Size (n in power of 2)")
plt.ylabel("Execution Time (ms)")
plt.title("Execution Time vs Array Size")
plt.xscale('log', base=2)  # Logarithmic scale for the x-axis (base 2)
plt.yscale('log')  # Logarithmic scale for the y-axis
plt.grid(True, which="both", linestyle='--', linewidth=0.5)

# Display the plot
plt.show()
```
