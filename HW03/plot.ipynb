## Task 3-1
Run task3 for value n = 106, value t = 8, and value ts = 21,22,··· ,210. Generate a plot called task3 ts.pdf which plots the time taken by your msort function vs. ts in linear–log scale.

```python
# Given data for time taken (in milliseconds) for different threshold values (ts)
# Given data for time taken (in milliseconds) for different threshold values (ts)
times_msort = [
    271.159, 163.274, 113.787, 61.729, 61.827, 54.9711, 47.8342, 47.734, 
    43.7178, 43.4282
]

# Adjust the thresholds to be powers of 2 (2, 4, 8, ..., 1024)
thresholds = [2**x for x in range(1, 11)]

# Plot the time vs threshold (ts) for the merge sort operation in linear-log scale
plt.figure(figsize=(8, 6))
plt.plot(thresholds, times_msort, marker='o', linestyle='-', color='r')
plt.xlabel("Threshold (ts)")
plt.ylabel("Time Taken (ms)")
plt.xscale("log", base=2)  # Set x-axis to log scale with base 2
plt.xticks(thresholds, labels=[str(t) for t in thresholds])  # Set custom labels to show 2, 4, 8, ..., 1024
plt.title("Time Taken by msort Function vs Threshold (ts) in Linear-Log Scale")
plt.grid(True)
plt.savefig('task3_ts_linear_log_with_labels.pdf')  # Save the plot as task3_ts_linear_log_with_labels.pdf

plt.show()
```
